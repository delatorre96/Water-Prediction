```python
import pandas as pd
import sqlite3
```

```python

# Crear conexión a la base de datos
conn = sqlite3.connect('BBDD/cantidadAguaJucar.db')

# Definir el número de registros por lote
batch_size = 100
offset = 0  # Inicializamos el offset en 0

# Crear un archivo CSV para guardar los datos
csv_file = 'datos_lotes.csv'
first_batch = True  # Flag para determinar si es el primer lote

# Iterar sobre los lotes de datos
while True:
    print(offset + batch_size)
    # Modificar la consulta SQL para incluir LIMIT y OFFSET
    query = f"""
    SELECT 
        df_copernicus.date_id,
        df_copernicus.'pixel',
        (SELECT rci.pixel 
         FROM df_rios_canales_id rci
         WHERE rci.pixel = pixeles_cercanos.pixel_rios_canales) AS pixel_rios_canales,
        (SELECT ai.pixel 
         FROM df_albufera_id ai
         WHERE ai.pixel = pixeles_cercanos.pixel_albufera) AS pixel_albufera,
        (SELECT eci.pixel 
         FROM df_embalses_cedex_id eci
         WHERE eci.pixel = pixeles_cercanos.pixel_embalses) AS pixel_embalses,
        df_copernicus.'latitude', df_copernicus.'longitude', df_copernicus.'total_precipitation',
        df_copernicus.'skin_temperature', df_copernicus.'evaporation', df_copernicus.'runoff', df_copernicus.'snowfall',
        df_copernicus.'soil_water_l1', df_copernicus.'soil_water_l2', df_copernicus.'soil_water_l3', df_copernicus.'soil_water_l4',
        df_copernicus.'high_vegetation_cover', df_copernicus.'low_vegetation_cover', df_copernicus.'type_high_vegetation',
        df_copernicus.'type_low_vegetation',
        (SELECT r.quantity_hm3
         FROM df_rios_canales r 
         JOIN df_rios_canales_id rci 
         ON rci.id_station = r.id_station 
         WHERE rci.pixel = pixeles_cercanos.pixel_rios_canales  AND r.date_id = df_copernicus.date_id) AS hm3_rios_canales,
         (SELECT a.quantity_hm3 
         FROM df_albufera a 
         JOIN df_albufera_id ai 
         ON ai.id_station = a.id_station 
         WHERE ai.pixel = pixeles_cercanos.pixel_albufera  AND a.date_id = df_copernicus.date_id) AS hm3_albufera,
        (SELECT e.quantity_hm3 
         FROM df_embalses_cedex e 
         JOIN df_embalses_cedex_id eci 
         ON eci.id_station = e.id_station 
         WHERE eci.pixel = pixeles_cercanos.pixel_embalses  AND e.date_id = df_copernicus.date_id) AS hm3_embalses
    FROM 
        df_copernicus
    JOIN pixeles_cercanos 
        ON df_copernicus.pixel = pixeles_cercanos.pixel_copernicus
    LIMIT {batch_size} OFFSET {offset};
    """

    # Leer los datos en un DataFrame
    df = pd.read_sql_query(query, conn)

    # Si no quedan más filas, salir del bucle
    if df.empty:
        break

    # Proceso de datos (asegúrate de manejar NAs y tipos de datos)
    

    # Guardar el DataFrame en el archivo CSV
    # Para el primer lote, escribimos el encabezado
    # Para los siguientes, solo añadimos los datos
    if first_batch:
        df.to_csv(csv_file, mode='w', index=False, header=True)
        first_batch = False
    else:
        df.to_csv(csv_file, mode='a', index=False, header=False)

    # Incrementar el offset para el siguiente lote
    offset += batch_size
# Cerrar la conexión a la base de datos
conn.close()
```
