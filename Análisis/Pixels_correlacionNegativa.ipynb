```python
import pandas as pd
from Func_extr import *
from Func_viz import *
from Func_analisis  import *     
import sqlite3
import folium
```

```python
conn = sqlite3.connect('BBDD/aguaCHJucar.db')

cursor = conn.cursor()

query = f'''
        SELECT location_id, latitude, longitude FROM locations_id WHERE Type = "Copernicus";
    '''

df_loc = pd.read_sql_query(query, conn)
# Ejecutar la consulta
cursor.execute(query)
conn.close()
```

```python
pixels_corrPos_D  = []
pixels_corrNeg_D  = []
pixels_corrPos_M  = []
pixels_corrNeg_M  = []
pixels_corrPos_Y  = []
pixels_corrNeg_Y  = []
for pixel in df_loc['location_id']:
    df = create_df(pixel)
    if df.empty:
        pass
    else:
        variables = ['total_precipitation']
        df = retardAvg_tNat(df, vars = variables, lags = [1,2,7,15,20], frec = 'D')
        df = retardAvg_tNat(df, vars = variables, lags = [1,2,3,4,5], frec = 'M')
        df = retardAvg_tNat(df, vars = variables, lags = [1,2,3,4,5], frec = 'Y')
        
        df = df.sort_values('date').set_index('date')
        df = df.select_dtypes(exclude=['datetime64'])
        df = df[[i for i in df.columns if 'precipitation' or i and 'quantity_hm3' in i]]
        correlaciones = df.corr()
        prec_columnsD = [i for i in correlaciones.columns if  'D' in i]
        media_corr = correlaciones[prec_columnsD].T['quantity_hm3'].mean()
        if media_corr>0:
            pixels_corrPos_D.append(pixel)
        else:
            pixels_corrNeg_D.append(pixel)
        prec_columnsM = [i for i in correlaciones.columns if 'M' in i]
        media_corr = correlaciones[prec_columnsM].T['quantity_hm3'].mean()
        if media_corr>0:
            pixels_corrPos_M.append(pixel)
        else:
            pixels_corrNeg_M.append(pixel)
        prec_columnsY = [i for i in correlaciones.columns if 'Y' in i]
        media_corr = correlaciones[prec_columnsY].T['quantity_hm3'].mean()
        if media_corr>0:
            pixels_corrPos_Y.append(pixel)
        else:
            pixels_corrNeg_Y.append(pixel)
df_corr_neg = df_loc[df_loc['location_id'].isin(pixels_corrNeg)]
```
