```python
df = pd.read_csv('df_copernicus.csv')
primera_fila = df.columns 
df.columns = range(df.shape[1])
primera_fila_dict = {}
for i in range(len( primera_fila)):
    primera_fila_dict[i] = primera_fila[i]
df_primeraFila = pd.DataFrame({727439  : primera_fila_dict}).T
df = pd.concat([df,df_primeraFila])
```

```python
nuevos_nombres = ['latitude', 'longitude', 'date', 'total_precipitation', 
                  'skin_temperature', 'evaporation', 'runoff', 'snowfall', 
                  'soil_water_l1', 'soil_water_l2', 'soil_water_l3', 'soil_water_l4',
                  'high_vegetation_cover', 'low_vegetation_cover', 
                  'type_high_vegetation', 'type_low_vegetation']

# AsegÃºrate de que el nÃºmero de nombres coincide con las columnas
df.columns = nuevos_nombres[:df.shape[1]]
```

```python
df.to_csv('df_copernicus_limpio.csv', index = False)
```
