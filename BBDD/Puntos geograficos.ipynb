```python
import xarray as xr
import pandas as pd
```

```python
ds = xr.open_dataset('Copernicus - climate ERA5/1960-01-01_a_1960-01-31.nc')
df = ds.to_dataframe()
df = df.reset_index()
df = df.drop(['number','expver'], axis = 1)
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Suponiendo que tu DataFrame se llama df y tiene columnas 'lat', 'lon'
# Reemplaza 'lat' y 'lon' con los nombres de tus columnas correspondientes
latitudes = df['latitude']
longitudes = df['longitude']

# Crear la figura
fig = plt.figure(figsize=(10, 5))

# Crear el eje usando la proyeccin de Mercator
ax = plt.axes(projection=ccrs.PlateCarree())

# Agregar caractersticas como costas y fronteras
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')

# Definir las extensiones del mapa (Espaa aproximadamente)
ax.set_extent([-10, 5, 35, 45], crs=ccrs.PlateCarree())  # [longitud oeste, longitud este, latitud sur, latitud norte]

# Graficar los puntos en el mapa
ax.scatter(longitudes, latitudes, color='red', s=10, transform=ccrs.PlateCarree())

# Ttulos
plt.title('Distribucin de puntos geogrficos en Espaa')
plt.show()
```

```python
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from shapely.geometry import Point
import geopandas as gpd

# Crear el eje usando la proyeccin PlateCarree
ax = plt.axes(projection=ccrs.PlateCarree())

# Cargar la caracterstica de masas terrestres (tierra)
land = cfeature.NaturalEarthFeature('physical', 'land', '10m')

# Convertir longitudes y latitudes en geometra de puntos
gdf_embalses = gpd.GeoDataFrame({
    'latitude': coord_df_embalses_id['latitude'],
    'longitude': coord_df_embalses_id['longitude']
}, geometry=gpd.points_from_xy(coord_df_embalses_id['longitude'], coord_df_embalses_id['latitude']))

# Crear una mscara de puntos que estn fuera de la tierra
# El feature de 'land' se toma como la representacin de tierra firme
land_geom = list(land.geometries())  # Geometras de la tierra firme

# Filtrar los embalses que estn fuera de tierra (en el mar)
embalses_en_mar = gdf_embalses[~gdf_embalses['geometry'].apply(lambda x: any([x.within(land_poly) for land_poly in land_geom]))]

# Mostrar los embalses que estn en el mar
print(embalses_en_mar)

# Mostrar visualizacin
plt.scatter(gdf_embalses['longitude'], gdf_embalses['latitude'], color='brown', s=10, label='Estaciones embalses')
plt.scatter(embalses_en_mar['longitude'], embalses_en_mar['latitude'], color='blue', s=20, label='Embalses en el mar')

# Ttulos y leyendas
plt.title('Embalses ubicados en el mar')
plt.legend()
plt.show()
```

```python
indexes = []
for index,row in coord_df_embalses_id.iterrows():
    for index2,row2 in embalses_en_mar.iterrows():
        if row['latitude'] == row2['latitude'] and row['longitude'] == row2['longitude']:
            indexes.append(index)
df_embalses_mar = df_embalses_id.loc[indexes]
```

```python
df_embalses_mar = df_embalses_id.loc[indexes]

def decimal_a_dms(coordenada, tipo='lat'):
    grados = int(coordenada)
    minutos_dec = abs(coordenada - grados) * 60
    minutos = int(minutos_dec)
    segundos = (minutos_dec - minutos) * 60
    
    # Determinar la direccin (N/S para latitudes, E/O para longitudes)
    if tipo == 'lat':
        direccion = 'N' if grados >= 0 else 'S'
    elif tipo == 'lon':
        direccion = 'E' if grados >= 0 else 'O'
    
    return f"{abs(grados)}{minutos}'{segundos:.2f}\" {direccion}"

df_embalses_mar['latitude_dms'] = df_embalses_mar['latitude'].apply(lambda x: decimal_a_dms(x, 'lat'))
df_embalses_mar['longitude_dms'] = df_embalses_mar['longitude'].apply(lambda x: decimal_a_dms(x, 'lon'))
```
