Como las coordenadas están el sistema de referencia UTM (Universal Transverse Mercator) usando el datum ETRS89 (European Terrestrial Reference System 1989)., vamos a cambiarlas por latitude y longitude de la misma manera que estñan en copernicus

```python
import pandas as pd
from pyproj import Proj, Transformer
```

```python
df_riosCanalespd = pd.read_csv('F2796_Rios_y_Canales_ROEA/F2796_D2_Serie día.csv', encoding='latin1',sep = ';')
df_identTotal = pd.read_csv('F2796_Rios_y_Canales_ROEA/F2796_M1_Actuales.csv', encoding='latin1',sep = ';')
df_identRio = pd.read_csv('F2796_Rios_y_Canales_ROEA/F2796_M2_Actuales_en_rio.csv', encoding='latin1',sep = ';')
df_identCanal = pd.read_csv('F2796_Rios_y_Canales_ROEA/F2796_M3_Actuales_en_canal.csv', encoding='latin1',sep = ';')
df_todos_id = pd.read_csv('F2796_Rios_y_Canales_ROEA/F2796_M0_Todas.csv', encoding='latin1', sep=';')
df_albufera_id = pd.read_excel('F56_Estaciones_de_aforo_de_L_Albufera/F56_M0_Estaciones_de_aforo_de_L_Albufera.xlsx')
```

```python
df_identTotal['UTM ETRS89 X'] = df_identTotal['UTM ETRS89 X'].str.replace(',', '.').astype(float)
df_identRio['UTM ETRS89 X'] = df_identRio['UTM ETRS89 X'].str.replace(',', '.').astype(float)
df_identCanal['UTM ETRS89 X'] = df_identCanal['UTM ETRS89 X'].str.replace(',', '.').astype(float)
df_todos_id['UTM ETRS89 X'] = df_todos_id['UTM ETRS89 X'].str.replace(',', '.').astype(float)
df_todos_id['UTM ETRS89 X'] = df_todos_id['UTM ETRS89 X'].str.replace(',', '.').astype(float)

df_identTotal['UTM ETRS89 Y'] = df_identTotal['UTM ETRS89 Y'].str.replace(',', '.').astype(float)
df_identRio['UTM ETRS89 Y'] = df_identRio['UTM ETRS89 Y'].str.replace(',', '.').astype(float)
df_identCanal['UTM ETRS89 Y'] = df_identCanal['UTM ETRS89 Y'].str.replace(',', '.').astype(float)
df_todos_id['UTM ETRS89 Y'] = df_todos_id['UTM ETRS89 Y'].str.replace(',', '.').astype(float)
df_todos_id['UTM ETRS89 Y'] = df_todos_id['UTM ETRS89 Y'].str.replace(',', '.').astype(float)
```

```python
proj_utm = Proj('EPSG:25830')
longitude = []
latitude = []
for index,row in df_identTotal.iterrows():
    lon, lat = proj_utm(row['UTM ETRS89 X'], row['UTM ETRS89 Y'], inverse=True)
    longitude.append(lon)
    latitude.append(lat)
df_identTotal['longitude'] = longitude
df_identTotal['latitude'] = latitude
longitude = []
latitude = []
for index,row in df_identRio.iterrows():
    lon, lat = proj_utm(row['UTM ETRS89 X'], row['UTM ETRS89 Y'], inverse=True)
    longitude.append(lon)
    latitude.append(lat)
df_identRio['longitude'] = longitude
df_identRio['latitude'] = latitude
longitude = []
latitude = []
for index,row in df_identCanal.iterrows():
    lon, lat = proj_utm(row['UTM ETRS89 X'], row['UTM ETRS89 Y'], inverse=True)
    longitude.append(lon)
    latitude.append(lat)
df_identCanal['longitude'] = longitude
df_identCanal['latitude'] = latitude
proj_utm = Proj('EPSG:25830')
longitude = []
latitude = []
for index,row in df_todos_id.iterrows():
    lon, lat = proj_utm(row['UTM ETRS89 X'], row['UTM ETRS89 Y'], inverse=True)
    longitude.append(lon)
    latitude.append(lat)
df_todos_id['longitude'] = longitude
df_todos_id['latitude'] = latitude
proj_utm = Proj('EPSG:25830')
longitude = []
latitude = []
for index,row in df_albufera_id.iterrows():
    lon, lat = proj_utm(row['UTM ETRS89 X'], row['UTM ETRS89 Y'], inverse=True)
    longitude.append(lon)
    latitude.append(lat)
df_albufera_id['longitude'] = longitude
df_albufera_id['latitude'] = latitude
```

```python
df_identTotal.to_csv('F2796_Rios_y_Canales_ROEA/F2796_M1_Actuales.csv')
df_identRio.to_csv('F2796_Rios_y_Canales_ROEA/F2796_M2_Actuales_en_rio.csv')
df_identCanal.to_csv('F2796_Rios_y_Canales_ROEA/F2796_M3_Actuales_en_canal.csv')
df_todos_id.to_csv('F2796_Rios_y_Canales_ROEA/F2796_M0_Todas.csv')
df_albufera_id.to_csv('F56_Estaciones_de_aforo_de_L_Albufera/F56_M0_Estaciones_de_aforo_de_L_Albufera.csv')
```
